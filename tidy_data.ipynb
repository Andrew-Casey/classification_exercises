Attendance Data

Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
attendance = pd.read_clipboard()
attendance
```

```python
attendance.student_id
```

```python
days_by_student = attendance.student[''].count()
```

```python
tardies_by_student = attendance_by_student['status'].apply(lambda x: (x == 'T').sum())
```

```python
half_days_by_student = attendance_by_student['status'].apply(lambda x: (x == 'H').sum() / 2)
```

```python
attendance_percentages = (days_by_student + half_days_by_student * 0.5 - tardies_by_student / 10) / days_by_student * 100


print(attendance_percentages)
```
